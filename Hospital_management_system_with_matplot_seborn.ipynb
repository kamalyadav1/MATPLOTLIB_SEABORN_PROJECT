```python
import pandas as pd
```

```python
import matplotlib.pyplot as plt
```

```python
import seaborn as sns
```

```python
df=pd.read_csv('hospital_management_long_dataset.csv')
```

```python
df.tail(50)
```

```python
dept_sum = df.groupby("Department")["Bill_Amount"].sum().reset_index()
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(8,5))

plt.bar(dept_sum["Department"], dept_sum["Bill_Amount"])

plt.xticks(rotation=45)
plt.xlabel("Department")
plt.ylabel("Total Earnings")
plt.title("Total Earnings by Department")
plt.show()
```
