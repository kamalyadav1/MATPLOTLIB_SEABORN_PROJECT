```python
import pandas as pd
import matplotlib.pyplot as plt
data = {
    "Business_Type": ["IT", "E-commerce", "Agro", "Healthcare", "Coaching",
                      "Retail", "IT", "Agro", "Healthcare", "Retail"],
    "Revenue": [12000000, 8000000, 6500000, 15000000, 7000000,
                4000000, 20000000, 9000000, 18000000, 3000000],
    "Profit_Margin": [35, 28, 25, 30, 40,
                      10, 45, 22, 33, 8],
    "Years_Running": [5, 4, 6, 7, 5,
                      2, 6, 4, 8, 1],
    "Growth_Rate": [18, 22, 15, 12, 20,
                    5, 25, 14, 16, 3],
    "Rating": [4.6, 4.4, 4.2, 4.7, 4.5,
               3.5, 4.8, 4.1, 4.6, 3.2]
}

```

```python
df = pd.DataFrame(data)
filtered_df = df[
    (df["Revenue"] > 5000000) &
    (df["Profit_Margin"] > 20) &
    (df["Years_Running"] > 3) &
    (df["Growth_Rate"] > 10) &
    (df["Rating"] > 4)
]

```

```python
success_count = filtered_df["Business_Type"].value_counts()
plt.figure(figsize=(8,5))

plt.bar(success_count.index,
        success_count.values,
        color=["red", "blue", "green", "purple", "orange"])

plt.title("Business Success Count by Type", fontsize=14)
plt.xlabel("Business Type")
plt.ylabel("Number of Successful Businesses")
plt.xticks(rotation=45)

plt.tight_layout()
plt.show()
```
