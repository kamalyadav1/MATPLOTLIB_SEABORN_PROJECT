```python
import pandas as pd
import matplotlib.pyplot as plt
data = {
    "Country": [
        "Switzerland","Norway","Ireland","Germany",
        "United States","China","Brazil","India",
        "Bangladesh","Nepal","Pakistan","Niger"
    ],
    "HDI_Value": [0.962,0.961,0.959,0.942,0.921,0.768,0.754,0.644,0.661,0.602,0.540,0.400],
    "Life_Expectancy": [84.0,83.2,82.3,81.0,77.2,78.2,72.8,67.2,72.4,68.5,66.1,61.6],
    "GNI_per_Capita": [66933,64660,76169,55340,69288,17504,14896,6590,5640,4137,4624,1240]
}

df = pd.DataFrame(data)
```

```python
plt.figure()
plt.bar(df["Country"], df["HDI_Value"])
plt.xticks(rotation=90)
plt.xlabel("Country")
plt.ylabel("HDI Value")
plt.title("HDI Value by Country (2022)")
plt.show()
```

```python
df_sorted = df.sort_values("HDI_Value")

plt.figure()
plt.plot(df_sorted["Country"], df_sorted["HDI_Value"])
plt.xticks(rotation=90)
plt.xlabel("Country")
plt.ylabel("HDI Value")
plt.title("HDI Comparison")
plt.show()
```
