```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
data = {
    "Vehicle_Type": ["Car", "Bike", "Truck", "Bus", "Auto", "Others"],
    "Accidents": [1200, 2500, 800, 600, 900, 400],
    "Deaths": [300, 900, 250, 180, 220, 100]
}

df = pd.DataFrame(data)
```

```python
plt.figure(figsize=(8,5))
plt.bar(df["Vehicle_Type"], df["Accidents"])
plt.title("Road Accidents by Vehicle Type")
plt.xlabel("Vehicle Type")
plt.ylabel("Number of Accidents")
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(8,5))
plt.bar(df["Vehicle_Type"], df["Deaths"])
plt.title("Road Deaths by Vehicle Type")
plt.xlabel("Vehicle Type")
plt.ylabel("Number of Deaths")
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(8,5))
plt.plot(df["Vehicle_Type"], df["Accidents"])
plt.plot(df["Vehicle_Type"], df["Deaths"])
plt.title("Accidents vs Deaths Comparison")
plt.xlabel("Vehicle Type")
plt.ylabel("Count")
plt.legend(["Accidents", "Deaths"])
plt.show()
```

```python
plt.figure(figsize=(6,6))
plt.pie(df["Deaths"], labels=df["Vehicle_Type"], autopct="%1.1f%%")
plt.title("Death Distribution by Vehicle Type")
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.barplot(x="Vehicle_Type", y="Deaths", data=df)
plt.title("Seaborn - Deaths by Vehicle")
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(5,4))
corr = df[["Accidents", "Deaths"]].corr()
sns.heatmap(corr, annot=True)
plt.title("Correlation Between Accidents and Deaths")
plt.show()
```
