```python
load_df=pd.read_csv('netflix_titles.csv')
```

```python
load_df.tail(5)
```

```python
load_df["type"].value_counts()
```

```python
count_table =load_df["type"].value_counts().reset_index()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(8,8))

colors = sns.color_palette("Set2", len(count_table))

plt.pie(
    count_table.iloc[:,1],
    labels=count_table.iloc[:,0],
    autopct='%1.1f%%',
    startangle=90,
    colors=colors,
    wedgeprops={'edgecolor':'white'}
)

plt.title("Movie vs TV Show", fontsize=15)
plt.axis('equal')
sns.set_theme(style="dark")
plt.show()
```
